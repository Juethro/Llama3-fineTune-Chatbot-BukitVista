```python
import pandas as pd
from datasets import Dataset
```

```python
data = pd.read_excel("dataset_bukit_vista.xlsx")
data.head()
```

```python
# Filter out rows where account.accountType is 'SERVICE'
data = data[data['account.accountType'] != 'SERVICE']

# Convert createdAtMs to datetime for sorting
data['createdAtMs'] = pd.to_datetime(data['createdAtMs'], unit='ms')
```

```python
data.isna().sum()
```

```python
# Drop rows where 'participantRole' is NaN/missing
data['participantRole'].fillna('GUEST', inplace=True)
data.head()
```

```python
data['participantRole'].unique()
```

```python
data.isna().sum()
```

```python
data = data.dropna(subset=['hydratedContent.body'])
data = data.groupby('threadId').apply(lambda x: x.sort_values('createdAtMs')).reset_index(drop=True)
```

## Convert to chatML (Template)

```python
def convert_to_chatml(thread_df):
    chat = []
    for _, row in thread_df.iterrows():
        role = "human" if row['participantRole'] == 'GUEST' else "gpt"
        message = {
            "from": role,
            "value": row['hydratedContent.body']
        }
        chat.append(message)
    return chat

# Mengelompokkan data berdasarkan threadId dan menerapkan fungsi konversi
threads = data.groupby('threadId').apply(convert_to_chatml).reset_index(name='chatml_content')
```

```python
# Slicing data for prototyping source code
# threads = threads.iloc[0:95]
```

## Clear Error Data

```python
from datasets import Dataset, concatenate_datasets

# Membuat dataset dari setiap elemen dalam threads['chatml_content']
data_error = []
datasets_list = []
for i, x in enumerate(threads['chatml_content']):
    try:
        datasets_list.append(Dataset.from_list(x))
    except TypeError :
        data_error.append(i)

print(data_error)
```

```python
# Mengeluarkan data yang indeksnya ada di data_error dari threads['chatml_content']
for index in sorted(data_error, reverse=True):
    del threads['chatml_content'][index]
```

```python
final = Dataset.from_dict({'conversations': threads['chatml_content']})
```
